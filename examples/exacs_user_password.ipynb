```python
import sys, os
# Adjust this if you've uploaded the plugin scripts/ dir elsewhere.
sys.path.insert(0, '/Workspace/Shared/oracle_ai_data_platform_connectors/scripts')
```

```python
from oracle_ai_data_platform_connectors.jdbc import build_oracle_jdbc_url, spark_jdbc_options_password

url = build_oracle_jdbc_url(
    host=os.environ['EXACS_HOST'],
    port=int(os.environ.get('EXACS_PORT_LEGACY', '1521')),
    service_name=os.environ['EXACS_SERVICE_NAME'],
    use_tcps=False,
)
opts = spark_jdbc_options_password(url=url, user=os.environ['EXACS_USER'], password=os.environ['EXACS_PASSWORD'])
```

```python
df = spark.read.format('jdbc').options(**opts).option('dbtable', os.environ['EXACS_TABLE_FOR_TEST']).load()
df.show(5)
```

```python
# Live-test driver parses this final cell's stdout for the JSON summary.
import json, time
summary = {
    'connector': 'aidp-exacs',
    'auth': 'password-legacy',
    'rows': int(df.count()),
    'schema': sorted([f.name for f in df.schema.fields]),
    'timestamp_utc': int(time.time()),
}
print('AIDP_LIVE_TEST_RESULT_BEGIN')
print(json.dumps(summary, indent=2))
print('AIDP_LIVE_TEST_RESULT_END')
```
